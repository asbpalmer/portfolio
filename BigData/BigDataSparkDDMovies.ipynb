**Part 1**: Fetch and preview the dataset

```python
i = -1
for header in data_list[0]:
    i += 1
    print(i, ") ", header)
```

We have country at index 7, but the data can't easily be processesd by line.split(','), because some items are lists

```python
def preserve_lists(line):
    sep_quotes = line.split(',')
    ret = []
    i = 0
    while i < len(sep_quotes):
        if len(sep_quotes[i]) > 0 and sep_quotes[i][0] == '"':
            temp = []
            while sep_quotes[i] and sep_quotes[i][-1] != '"':
                temp.append(sep_quotes[i][1:])
                i += 1
            temp.append(sep_quotes[i][:-1])
            i += 1
            ret.append(temp)
        else:
            ret.append([sep_quotes[i]])
            i += 1
    return ret
```

```python
dataset = data.collect()
```

```python
for i in range(len(dataset)):
    c = preserve_lists(dataset[i])[7]
    if len(c) > 1:
        print([country.strip() for country in c])
```

A lot of movies have more than 1 country, so we have to handle that

**Part 2**: Map the dataset to (country, 1) and reduce by the key

```python
countries = data.flatMap(lambda line: [(c.strip(),1) for c in preserve_lists(line)[7]])
```

```python
countries.collect()
```

```python
countries_counts = countries.reduceByKey(lambda a, b: a+b)
```

```python
countries_counts.collect()
```

**Part 3**: Create Bar Chart

```python
def create_bar_chart(line):
    bar = ''
    thousands = line[1]//1000
    return (line[0], f"{bar*thousands}({line[1]/1000:.2f}k)")
```

```python
def remove_null(line):
    if line[0]:
        return line
    else:
        return ("N/A", line[1])
```

```python
final = countries_counts.map(create_bar_chart).map(remove_null).sortBy(lambda line: line[1], ascending=False)
```

**Part 4**: Saving and Showing

```python
import os
if os.path.exists('results'):
    os.rmdir('results')
final.saveAsTextFile('results')
```

```python
for r in final.collect():
    if r:
        print(r)
```
